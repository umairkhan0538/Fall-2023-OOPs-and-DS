    <strong style="color: #1abc9c; background-color: #e8f8f5; padding: 12px; border-radius: 8px; display: block; margin-bottom: 10px;">Drop Some Unnecessary Columns</strong>
        <strong style="color: #e74c3c;">Removing unnecessary columns helps simplify the dataset.</strong><br>
        <span style="color: #2d3436;">To focus on relevant features, we can drop columns that do not contribute to our analysis or model performance.
```python
Data = Data.drop('date',axis=1)
Data = Data.drop('id',axis=1)
Data = Data.drop('zipcode',axis=1)
```

<ol style="color: #2c3e50; font-family: 'Arial', sans-serif; line-height: 1.8; padding: 10px;">
    <strong style="color: #16a085; background-color: #eaf7f6; padding: 12px; border-radius: 8px; display: block; margin-bottom: 10px;">Step 2: Dataset Preparation (Splitting and Scaling)</strong>
        <strong style="color: #e74c3c;">Data is divided into the Train set and Test set.</strong><br>
        <span style="color: #34495e;">We use the Train set to make the algorithm learn the data’s behavior and then check the accuracy of our model on the Test set.</span><br>
        <span style="color: #34495e;">Features (X): The columns that are inserted into our model will be used to make predictions.</span><br>
        <span style="color: #34495e;">Prediction (y): Target variable that will be predicted by the features.</span>
    </li>
</ol>


```python
Data.dropna(inplace=True)
```

```python
X = Data.drop('price',axis =1).values
y = Data['price'].values
```

```python
#splitting Train and Test 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
#standardization scaler - fit&transform on train, fit only on test

s_scaler = StandardScaler()
X_train = s_scaler.fit_transform(X_train)
X_test = s_scaler.transform(X_test)
```

<ol style="color: #2c3e50; font-family: 'Arial', sans-serif; line-height: 1.8; padding: 10px;">
    <strong style="color: #e67e22; background-color: #fdf2e6; padding: 12px; border-radius: 8px; display: block; margin-bottom: 10px;">Step 3: Model Selection and Evaluation</strong>
        <strong style="color: #c0392b;">💡 Model 1: Multiple Linear Regression</strong><br>
        <span style="color: #2c3e50;">Multiple Linear Regression is an extension of Simple Linear Regression and assumes that there is a linear relationship between a dependent variable <em>Y</em> and independent variables <em>X</em>.</span><br>
        <span style="color: #2c3e50;">Let’s wrap the training process in our Regression model:</span>
    </li>
</ol>


```python
# Multiple Liner Regression

regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
#evaluate the model (intercept and slope)
print(regressor.intercept_)
print(regressor.coef_)
```

```python
#predicting the test set result
y_pred = regressor.predict(X_test)
```

```python
#put results as a DataFrame
coeff_df = pd.DataFrame(regressor.coef_, Data.drop('price',axis =1).columns, columns=['Coefficient']) 
coeff_df
```

```python
# visualizing residuals
fig = plt.figure(figsize=(10,5))
residuals = (y_test- y_pred)
sns.distplot(residuals)
```

<ol style="color: #2c3e50; font-family: 'Arial', sans-serif; line-height: 1.8; padding: 10px;">
    <strong style="color: #8e44ad; background-color: #f4e7fa; padding: 12px; border-radius: 8px; display: block; margin-bottom: 10px;">Compare Actual Output and Predicted Value</strong>
        <strong style="color: #e74c3c;">Let’s compare actual output and predicted value to measure how far our predictions are from the real house prices.</strong><br>
    </li>
</ol>


```python
#compare actual output values with predicted values
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(10)
df1
```

```python
# evaluate the performance of the algorithm (MAE - MSE - RMSE)

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```

<ol style="color: #2c3e50; font-family: 'Arial', sans-serif; line-height: 1.8; padding: 10px;">
    <strong style="color: #1abc9c; background-color: #eafaf1; padding: 12px; border-radius: 8px; display: block; margin-bottom: 10px;">💡 Model 2: Keras Regressions</strong>
        <strong style="color: #c0392b;">Let’s create a baseline neural network model for the regression problem.</strong><br>
        <span style="color: #34495e;">Starting with all of the needed functions and objects.</span>
    </li>
</ol>


<ol style="color: #2c3e50; font-family: 'Arial', sans-serif; line-height: 1.8; padding: 10px;">
    <strong style="color: #1abc9c; background-color: #eafaf1; padding: 12px; border-radius: 8px; display: block; margin-bottom: 10px;">Model Architecture</strong>
        <strong style="color: #c0392b;">Since we have 19 features, let’s insert 19 neurons as a start, 4 hidden layers and 1 output layer to predict house price.</strong><br>
        <span style="color: #34495e;">Also, ADAM optimization algorithm is used for optimizing the loss function (Mean Squared Error).</span>
    </li>
</ol>


```python
# having 19 neuron is based on the number of available features
model = Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='Adam',loss='mes')
```

<ol style="color: #2c3e50; font-family: 'Arial', sans-serif; line-height: 1.8; padding: 10px;">
    <strong style="color: #1abc9c; background-color: #eafaf1; padding: 12px; border-radius: 8px; display: block; margin-bottom: 10px;">Training the Model</strong>
        <strong style="color: #c0392b;">Then, we train the model for 50 epochs, and each time record the training and validation accuracy in the history object.</strong><br>
        <span style="color: #34495e;">To keep track of how well the model is performing for each epoch, the model will run on both train and test data while also calculating the loss function.</span>
    </li>
</ol>


```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=50)
model.summary()
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot(figsize=(12,8))
```

<ol style="color: #2c3e50; font-family: 'Arial', sans-serif; line-height: 1.8; padding: 10px;">
    <strong style="color: #8e44ad; background-color: #f4e7fa; padding: 12px; border-radius: 8px; display: block; margin-bottom: 10px;">Evaluation on Test Data</strong>
    </li>
</ol>


```python
y_pred = model.predict(X_test)

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```

```python
# Visualizing Our predictions
fig = plt.figure(figsize=(10,5))
plt.scatter(y_test,y_pred)
# Perfect predictions
plt.plot(y_test,y_test,'r')
```

```python
# visualizing residuals
fig = plt.figure(figsize=(10,5))
residuals = (y_test- y_pred)
sns.distplot(residuals)
```
